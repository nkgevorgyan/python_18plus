# Знакомство с Git и Github

## 1. Что такое Git.

**Git** - это бесплатная распределенная система управления версиями с открытым исходным кодом, предназначенная для работы как с небольшими, так и с крупными проектами.

Контроль версий относится к процессу сохранения различных файлов или «версий» на различных этапах проекта. Это позволяет разработчикам отслеживать, что было сделано, и возвращаться к предыдущему этапу, если они решат отменить некоторые из внесенных изменений.

Это полезно по разным причинам. Например, это упрощает устранение ошибок и исправление других ошибок, которые могут возникнуть во время разработки. Вы также можете отмечать изменения в каждой версии, чтобы помочь всем членам команды оставаться в курсе того, что выполнено, а что еще нужно сделать.

В отличие от большинства других систем управления версиями, **git** хранит каждую сохраненную версию как «снимок» вместо списка изменений, внесенных в каждый файл. Вы можете ссылаться на старые снимки, когда вам нужно, и новые снимки создаются при изменении вашего проекта.

**Git** также позволяет вам «отправлять» и «извлекать» изменения в установки на других компьютерах и обратно. Это делает ее так называемой «распределенной системой контроля версий» и позволяет нескольким разработчикам работать над одним и тем же проектом.

## 2. В чем разница между Git и Github.

GitHub упрощает совместную работу с помощью git. Это платформа, которая может хранить репозитории кода в облачном хранилище, чтобы несколько разработчиков могли работать над одним проектом и видеть изменения друг друга в режиме реального времени:

Кроме того, он также включает в себя функции организации и управления проектами. Вы можете назначать задачи отдельным лицам или группам, устанавливать разрешения и роли для соавторов, а также использовать модерацию комментариев, чтобы держать всех под контролем.

Публичные репозитории GitHub доступны доступны для каждого. Разработчики со всего мира могут взаимодействовать с кодом друг друга и вносить в него свой вклад, чтобы изменять или улучшать его, что называется «социальным кодированием». В некотором смысле это делает GitHub социальной сетью для разработчиков.

Когда дело доходит до взаимодействия с кодом других разработчиков на GitHub, вы можете предпринять три основных действия:

* **Fork**: процесс копирования чужого кода из репозитория с целью его изменения.
* **Pull**: когда вы закончите вносить изменения в чужой код, вы можете поделиться ими с первоначальным владельцем с помощью **pull request**.
* **Merge**: владельцы могут добавлять новые изменения в свои проекты посредством слияния и отдавать должное участникам, которые их предложили.

**git** - это локальное программное обеспечение для  управления версиями, которое позволяет разработчикам сохранять моментальные снимки своих проектов с течением времени. Как правило, лучше всего подходит для индивидуального использования.

**GitHub** - это веб-платформа, которая включает в себя функции управления версиями git, поэтому их можно использовать совместно. Он также включает в себя функции управления проектами и командами, а также возможности для работы в сети и социального программирования.

## 3. Создание первого репозитория

**Шаг 1:** Сначала загрузите программное обеспечение **git** для своей операционной системы

![Github_log](images/download-git.png)

Затем вам нужно будет запустить установщик. Этот процесс будет зависеть от того, какую ОС вы используете. **Git** можно настроить используя интерфейс командной строки.

*Есть несколько терминов, с которыми нужно ознакомиться:*

* **Repository**: расположение файла, в котором хранится ваш проект.

* **Commit**: команда, используемая для сохранения новых изменений вашего проекта в репозитории.
* **Stage**: прежде чем вы сможете сохранить изменения в Git, вам необходимо сначала зафиксировать их - это дает вам возможность подготовить код перед его официальным добавлением в проект.
* **Branch**: часть вашего проекта, которую вы активно развиваете.

Чтобы подключить **git** к **GitHub**, вам нужно добавить репозиторий и сделать хотя бы один **commit**. После этого у вас будет достаточно созданного проекта, чтобы начать работу в **GitHub**.

 **Шаг 2**: Создать **Github** аккаунт
 
Затем вам понадобится учетная запись GitHub. Вы можете зарегистрироваться и выбрать бесплатный план

![Github_log](images/github-pricing.png)

**Шаг 3**: Добавьте репозиторий **GitHub** в свою учетную запись

После того, как вы создали и настроили свою учетную запись, вам нужно будет создать репозиторий в GitHub, где вы сможете хранить свой проект, когда вы переместите его из **git**. Вы можете сделать это, щелкнув соответствующую ссылку на левой боковой панели:

![Github_log](images/create-github-repository.png)

Затем вам нужно будет выбрать имя для вашего репозитория:

![Github_log](images/new-repository-name.png)

Вы также можете решить, хотите ли вы, чтобы репозиторий был общедоступным, чтобы другие пользователи **GitHub** могли его находить и вносить в него изменения, или если вы хотите, чтобы он оставался частным.

**Шаг 4**: Сделать **Push** репозитория в **GitHub**


Затем у вас будет возможность добавить код в репозиторий несколькими способами.

Поскольку вы уже настроили свой репозиторий **git**, вы можете использовать принудительную отправку существующего репозитория из параметра командной строки:

![Github_log](images/push-existing-repository.png)

Значок буфера обмена справа позволяет скопировать перечисленные здесь команды, чтобы вы могли быстро вставить их в предпочитаемый интерфейс командной строки для создания репозитория **GitHub**.

Как только это будет сделано, обновите страницу **GitHub**.


**Шаг 5**: Сделайте **Pull** своих изменений в **git**

Вы можете увидеть все изменения, которые вы и другие внесли в свой проект, на GitHub, но платформа не имеет прямого доступа к файлам вашего компьютера. Чтобы поддерживать проект в актуальном состоянии на вашем компьютере, вам нужно будет получить свои правки через git.

Для этого просто введите **git pull origin master** в интерфейс командной строки. Это должно обновить ваши файлы, чтобы все было синхронизировано на всех итерациях вашего проекта.
