## 2. Типы данных в Python


### Переменная

Переменные - это не что иное, как зарезервированные ячейки памяти для хранения значений. Это значит, что когда
вы создаете переменную, вы резервируете место в памяти. 

В зависимости от типа данных переменной интерпретатор выделяет память и решает, что можно сохранить в зарезервированной памяти. Следовательно, назначая различные типы данных для переменных, вы можете хранить в этих переменных целые числа, десятичные дроби или символы.

### Присвоение значений переменным

Переменные Python не нуждаются в явном объявлении для резервирования места в памяти. Объявление
происходит автоматически, когда вы присваиваете значение переменной. Знак равенства (=) используется для того, чтобы присвоить значения переменным.

Операнд слева от оператора (=) - это имя переменной, и операнд справа от оператора (=) - это значение, хранящееся в переменной.

```python
age = 10                    # Объявление целого числа - integer

average_mark = 8.6          # Объявление числа с плавающей точкой - float

name = 'Вася'               # Объявление строки - string
```

### Числа в Python

Числовые типы данных хранят числовые значения. Числовые объекты создаются, когда вы назначаете для них значение.

```python
number1 = 1             # Объявление целого числа - int

number2 = 10.8          # Объявление дробного числа - float

number3 = 3 + 5j        # Объявление комплексного числа - integer
```

Python поддерживает три разных числовых типа
* **int** (целые числа со знаком)
* **float** (вещественные значения с плавающей запятой)
* **комплексные** (комплексные числа)

### Строки в Python

Строки в Python идентифицируются как последовательный набор символов, представленных в
кавычках. Python допускает использование пары одинарных или двойных кавычек. Подмножества строки
можно рассмотреть с помощью оператора среза ([] и [:]) с индексами, начинающимися с 0 в
начале строки и до -1 в конце.

Знак плюс (+) - это оператор конкатенации строк, а звездочка (*) - оператор повторения

```python
my_string = 'А я строка'

print (my_string)           # Выводится строка
print (my_string[0])        # Выводится первый символ в строке
print (my_string[2:5])      # Выводятся символы с 3ей по 5ую
print (my_string[2:])       # Выводится подстрока начианаю с 3его символа
print (my_string * 2)       # Выводит строку два раза
print (my_string + "TEST")  # Выводит объединенную строку
```
